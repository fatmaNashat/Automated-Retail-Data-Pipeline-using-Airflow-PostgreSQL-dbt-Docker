```python

from sqlalchemy import create_engine, text
engine = create_engine('postgresql://postgres:123@localhost:5432/retail_db')

with engine.connect() as conn:
    trans = conn.begin()
    conn.execute(text("TRUNCATE TABLE raw.walmart_sales"))
    trans.commit()
```
